```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
""" Your code hear: 讀取 train.csv, test.csv 並合併 """
train=pd.read_csv('../train.csv')
test=pd.read_csv('../test.csv')
df=pd.concat([train,test], axis = 0)
```

```python
# 將目標分析欄位取出來
goal=df[['LotFrontage', 'GarageYrBlt', 'MasVnrArea']]
```

```python

missing = goal.isnull().sum(axis = 0)
missing = missing[missing > 0]
missing.sort_values(inplace=True, ascending = False)

plt.figure(figsize = (8, 4))
plt.bar(list(missing.index), np.array(missing))
plt.xticks(rotation='vertical')
plt.show()
```

```python
L=df['LotFrontage'].copy()
L_mean=L.copy();L_median=L.copy()
L_mean.fillna(value=L.mean(), inplace=True)
L_median.fillna(value=L.median(), inplace=True)

G=df['GarageYrBlt'].copy()
G_mean=G.copy();G_median=G.copy()
G_mean.fillna(value=G.mean(), inplace=True)
G_median.fillna(value=G.median(), inplace=True)

M=df['MasVnrArea'].copy()
M_mean=M.copy();M_median=M.copy()
M_mean.fillna(value=M.mean(), inplace=True)
M_median.fillna(value=M.median(), inplace=True)
```

```python
fig = plt.figure(figsize = (10, 5))

ax1 = plt.subplot(1, 3, 1, title = 'first')
X=[df['LotFrontage'],L_mean,L_median]
sns.histplot(data =df['LotFrontage'] , ax = ax1,kde=True,label='orgin')
sns.histplot(data=L_mean,ax=ax1,kde=True,color='y',label='mean')
sns.histplot(data=L_median,ax=ax1,kde=True,color='r',label='median')
plt.legend()
ax2 = plt.subplot(1, 3, 2, title = 'second')
sns.histplot(data = df['GarageYrBlt'], kde = True, ax = ax2,label='orgin')
sns.histplot(data=G_mean,ax=ax2,kde=True,color='y',label='mean')
sns.histplot(data=G_median,ax=ax2,kde=True,color='r',label='median')
plt.legend()
ax3 = plt.subplot(1, 3, 3, title = 'third')
sns.histplot(data = df['MasVnrArea'], kde = True, ax = ax3,label='orgin')
sns.histplot(data=M_mean,ax=ax3,kde=True,color='y',label='mean')
sns.histplot(data=M_median,ax=ax3,kde=True,color='r',label='median')
plt.legend()
```

```python
print(f'LotFrontage 標準差: \n原值:{L.std()}, \n填平均值:{L_mean.std()}, \n填中位數:{L_median.std()}')
```

```python
print(f'GarageYrBlt 標準差: \n原值:{G.std()}, \n填平均值:{G_mean.std()}, \n填中位數:{G_median.std()}')
```

```python
print(f'MasVnArea 標準差: \n原值:{M.std()}, \n填平均值:{M_mean.std()}, \n填中位數:{M_median.std()}')
```
